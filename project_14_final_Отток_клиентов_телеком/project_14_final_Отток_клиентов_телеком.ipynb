```python
def df_info(dataframe):
    column = dataframe.columns.values.tolist() # список наименований столбцов таблицы
    print('Общая информация о таблице:\n')
    dataframe.info() # общая иноформация о таблице
    print('\nВнешний вид таблицы (5 случайных строк):')
    display(dataframe.sample(5)) # внешний вид первых 5 строк таблицы

    # далее выводим все интересующие нас особенности таблицы, соответствнно, обозначенные в функции print()
    print('Столбцы таблицы: {}\n'.format(column),
          '\n\nОбщее количество уникальных значений столбцов {}:\n\n{}'.
          format(column, dataframe.nunique()),
          '\n\nКоличество пропущенных значений столбцов {}:\n\n{}'.
          format(column, dataframe.isna().sum()))
    
    for i in column: # создадим цикл для перебора значений столбцов для отдельного примнения к ним любой функции
          print('\n\nСоотношение дубликатов столбца {} ко всем строкам:\t {}'.format(i, dataframe[i].duplicated().mean()))
    print('\nСоотношение явных дубликтов ко всем строкам: {}'.format((dataframe.duplicated().mean())))
    print('\nСоотношение пропусков ко всем строкам: {}'.format((dataframe.isna().mean())))
    print('\n\nТаблица корреляций значений столбцов {}:\n\n{}'.format(column, dataframe.corr()), '\n',
          '\n\nТаблица распределений значений столбцов {}:\n\n{}'.format(column, dataframe.describe()), '\n')
    
```

Составим список признаков, из которых будем создавать таблицу

```python
features_list = ['customerID', 'BeginDate', 'EndDate', 'Type', 'PaperlessBilling', 
                 'PaymentMethod', 'MonthlyCharges', 'TotalCharges', 
                 'MultipleLines', 'InternetService','OnlineSecurity',
                 'OnlineBackup','DeviceProtection','TechSupport',
                 'StreamingTV','StreamingMovies']
```

```python
data_features = df_contract[['customerID','BeginDate', 'EndDate', 'Type', 'PaperlessBilling', 
                 'PaymentMethod', 'MonthlyCharges', 'TotalCharges']]
```

```python
data_features = data_features.merge(df_phone, on=['customerID'], how='outer')
```

```python
data_features = data_features.merge(df_internet, on=['customerID'], how='outer')
```

```python
data_features.head()
```

```python
data_features['TotalCharges'] = pd.to_numeric(data_features['TotalCharges'], errors='coerce')
```

```python
data_features.isna().sum()
```

```python
data_features.dropna(subset=['TotalCharges'], inplace=True)
```

```python
data_features.isna().sum()
```

Так как некоторые пользователи не пользуются многими сервисами, которыми и являются ячейки с пропусками, можем просто заменить их на значение `No`

```python
data_features = data_features.fillna('No')
```

```python
data_features.columns = data_features.columns.str.lower()
```

```python
data_features.head()
```

```python
categoric = ['enddate', 'type', 'paperlessbilling', 'paymentmethod', 
             'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup',
             'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies']
```

```python
for i in categoric:
    data_features[i] = data_features[i].str.lower()
```

```python
data_features['begindate'] = data_features['begindate'].astype('datetime64')

data_features.loc[data_features['enddate'] == 'no', 'enddate'] = '2020-02-01'
data_features['enddate'] = data_features['enddate'].astype('datetime64')
data_features['begin_days'] = (pd.Timestamp('2020-02-01') - data_features['begindate']) // np.timedelta64(1, 'D')
data_features['end_days'] = (pd.Timestamp('2020-02-01') - data_features['enddate']) // np.timedelta64(1, 'D')
data_features['liveclient'] = (data_features['end_days'] > 0).astype('int')

data_features['lifetime'] = data_features['begin_days'] - data_features['end_days']
```

## Разбиение данных на выборки

Выделим отдельно таблицу целевого признака и сделаем две разные обработки таблицы признаков

```python
features = features_drop.drop(['liveclient'], axis=1)
```

```python
target = features_drop['liveclient']
```

```python
numeric = ['monthlycharges', 'lifetime', 'totalcharges']
cat_features = [col for col in list(features.columns) if col not in numeric]
cat_features
```

```python
features.sample(5)
```

Создадим первую таблицу с признаками и обработаем ее с помощью модуля `dummy`

```python
features_ohe = pd.get_dummies(features, drop_first=False)
```

```python
features_ohe.sample(5)
```

Используем и другую кодировку признаков - порядковую

Для начала создадим список категориальных переменных

Создадим новый датафрейм, чтобы не трогать исходный

```python
features_encodered = features.copy()
```

```python
encoder = OrdinalEncoder() # создадим модель стандартизации
encoder.fit(features_encodered[cat_features]) # и обучим ее на необходимых признаках

# заменим исходные значения признаков на стандартизированные
features_encodered[cat_features] = encoder.transform(features_encodered[cat_features])
```

```python
features_encodered.head()
```

**Выводы**: таблицы подготовлены для дальнейшей работы. Приступим

С порядковым кодированием

```python
features_encodered_train, features_encodered_test, target_encodered_train, target_encodered_test = train_test_split(
    features_encodered, target, test_size=0.25, random_state=RS)
```

```python
print('Размер обучающей выборки: ', features_encodered_train.shape, 
      'Размер тестовой выборки: ', features_encodered_test.shape, sep='\n')
```

```python
features_encodered_train.head()
```

```python
features_encodered_train.corr()
```

**Выводы**: таким образом, наши выборки подготовлены для дальнейшего обучения моделей

В силу приемлемых значений корреляций признаков будем использовать для дальнейшей работы таблицу с признаками, обработанными порядковым способом, чтобы избежать мультиколлинеарности.

Чтобы избежать ошибки разномасштабных признаков, необходимо применить стандартизацию данных нужных нам столбцов. Сделаем это заранее

```python
scaler_encodered = StandardScaler() # создадим модель стандартизации
scaler_encodered.fit(features_encodered_train[numeric]) # и обучим ее на тренировочных признаках
pd.options.mode.chained_assignment = None # избавимся от типового предупреждения

# заменим значения тренировочных, валидных и тестовых признаков на стандартизированные
features_encodered_train[numeric] = scaler_encodered.transform(features_encodered_train[numeric]) 
features_encodered_test[numeric] = scaler_encodered.transform(features_encodered_test[numeric])
```

```python
features_encodered_train.head()
```

```python
features_encodered_train.corr()
```

## Обучение моделей

```python
model_lr = LogisticRegression(random_state=RS).fit(features_encodered_train, target_encodered_train)
probs_lr = model_lr.predict_proba(features_encodered_test)[:, 1]

model_dt = DecisionTreeClassifier(random_state=RS).fit(features_encodered_train, target_encodered_train)
probs_dt = model_dt.predict_proba(features_encodered_test)[:, 1]

model_rf = RandomForestClassifier(random_state=RS, n_jobs= -1).fit(features_encodered_train, target_encodered_train)
probs_rf = model_rf.predict_proba(features_encodered_test)[:, 1]

model_xg = XGBClassifier(random_state=RS).fit(features_encodered_train, target_encodered_train)
probs_xg = model_xg.predict_proba(features_encodered_test)[:, 1]

# model_lgb = LGBMClassifier(random_state=RS, n_jobs= -1).fit(features_encodered_train, target_encodered_train)
# probs_lgb = model_lgb.predict_proba(features_encodered_test)[:, 1]

model_cb = CatBoostClassifier(random_state=RS).fit(features_encodered_train, target_encodered_train)
probs_cb = model_cb.predict_proba(features_encodered_test)[:, 1]
```

```python
auc_lr = roc_auc_score(target_encodered_test, probs_lr)
fpr_lr, tpr_lr, thresholds_lr = roc_curve(target_encodered_test, probs_lr)

auc_dt = roc_auc_score(target_encodered_test, probs_dt)
fpr_dt, tpr_dt, thresholds_dt = roc_curve(target_encodered_test, probs_dt)

auc_rf = roc_auc_score(target_encodered_test, probs_rf)
fpr_rf, tpr_rf, thresholds_rf = roc_curve(target_encodered_test, probs_rf)

auc_xg = roc_auc_score(target_encodered_test, probs_xg)
fpr_xg, tpr_xg, thresholds_xg = roc_curve(target_encodered_test, probs_xg)

# auc_lgb = roc_auc_score(target_encodered_test, probs_lgb)
# fpr_lgb, tpr_lgb, thresholds_lgb = roc_curve(target_encodered_test, probs_lgb)

auc_cb = roc_auc_score(target_encodered_test, probs_cb)
fpr_cb, tpr_cb, thresholds_cb = roc_curve(target_encodered_test, probs_cb)

plt.figure(figsize=(12, 7))
plt.plot(fpr_lr, tpr_lr, label=f'AUC (Logistic Regression) = {auc_lr:.2f}')
plt.plot(fpr_dt, tpr_dt, label=f'AUC (Decision Tree) = {auc_dt:.2f}')
plt.plot(fpr_rf, tpr_rf, label=f'AUC (Random Forests) = {auc_rf:.2f}')
plt.plot(fpr_xg, tpr_xg, label=f'AUC (XGBoost) = {auc_xg:.2f}')
# plt.plot(fpr_lgb, tpr_lgb, label=f'AUC (LGBMClassifier) = {auc_lgb:.2f}')
plt.plot(fpr_cb, tpr_cb, label=f'AUC (CatBoostClassifier) = {auc_cb:.2f}')

plt.plot([0, 1], [0, 1], color='blue', linestyle='--', label='Baseline')
plt.title('ROC Curve', size=20)
plt.xlabel('False Positive Rate', size=14)
plt.ylabel('True Positive Rate', size=14)
plt.legend();
```

```python
final_encodered = pd.DataFrame({'model name': [model_lr.__class__.__name__,
                                     model_dt.__class__.__name__,
                                     model_rf.__class__.__name__,
                                     model_xg.__class__.__name__,
#                                      model_lgb.__class__.__name__, 
                                     model_cb.__class__.__name__],
                                    'Значение ROC-AUC': [auc_lr,
                                                         auc_dt,
                                                         auc_lr,
                                                         auc_xg,
#                                                          auc_lgb,
                                                         auc_cb]}).sort_values(
    by='Значение ROC-AUC', ascending=False).reset_index(drop=True).reset_index(drop=True)
display('Таблица итоговых значений ROC-AUC обученных моделей', final_encodered)
```

```python
importances = model_cb.feature_importances_

# train_features is the dataframe of training features
feature_list = list(features_encodered_test.columns)

# Extract the feature importances into a dataframe
feature_results = pd.DataFrame({'feature': feature_list,'importance': importances})

# Show the top 10 most important
feature_results = feature_results.sort_values('importance',ascending = False).reset_index(drop=True)

display('Таблица значимости признаков для модели {}'.format(model_cb.__class__.__name__),feature_results.head(10))
```

2. Довольно непросто было добавить новые целевые признаки, что потребовало проработки двух вариантов его получения (получения за счет месяцев оплаты и получение дополнительного признаа "жизни клиента" по дням пользования телеком сервиса)
3. Сложности были со скоростью обучения различных моделей, особенно при кросс-валидизации признаков, что требовало много времени.
